## FILTROS Y HERRAMIENTAS PARA EL PRE-PROCESO DE IMÁGENES DIGITALES

En el análisis de imágenes, los filtros y otras herramientas juegan un papel crucial para mejorar, procesar e interpretar imágenes, especialmente cuando se trabaja con datos en bruto en investigaciones científicas, diagnósticos médicos o aplicaciones de visión por computadora. A continuación, se presenta una descripción general de los tipos comunes de filtros y herramientas, y su importancia en el análisis de imágenes:

# FILTROS ESCENCIALES
## 1. Filtros de Suavizado (Desenfoque)
- **Tipos**: Filtro Gaussiano (gaussian_blur), filtro de mediana (median_blur), filtro de promedio
- **Propósito**: Estos filtros reducen el ruido y los detalles en una imagen. A menudo se utilizan en el preprocesamiento para eliminar ruido aleatorio mientras se preservan características importantes, como los bordes.
- **Importancia**: El suavizado ayuda a mejorar la claridad de las características clave de la imagen, lo que facilita el análisis posterior, como la detección de objetos o la segmentación.

## 2. Filtros de Enfoque (Afilado)
- **Tipos**: Filtro Laplaciano (laplace_filter), filtro de máscara de desenfoque, filtro pasa-altos
- **Propósito**: El enfoque mejora los bordes y los detalles finos al resaltar áreas de cambio rápido de intensidad.
- **Importancia**: Estos filtros ayudan a enfatizar características importantes, facilitando la detección de límites y estructuras en la imagen.

## 3. Filtros de Detección de Bordes
- **Tipos**: Filtro Sobel (sobel_filter), detector de bordes Canny(Canny_edge_detect), filtro Prewitt
- **Propósito**: Se utilizan para identificar puntos en una imagen donde la intensidad de los píxeles cambia significativamente, lo que a menudo corresponde a los bordes de objetos o características clave.
- **Importancia**: Crucial para el reconocimiento de objetos, segmentación y extracción de características. Los bordes ayudan a definir formas y son esenciales para comprender la estructura de los objetos en una imagen.

## 4. Umbralización (Thresholding)
- **Tipos**: Global, adaptativo, binario (threshold_binary), método de Otsu (threshold_otsu), textura (theshold_texture),  gaussiano (threshold_gaussian) y otros
- **Propósito**: La umbralización convierte imágenes en escala de grises en imágenes binarias, donde los píxeles se clasifican como primer plano o fondo en función de un umbral de intensidad de píxeles.
- **Importancia**: A menudo se utiliza para segmentar objetos del fondo, lo que simplifica el análisis de áreas específicas de interés en una imagen.

# FILTROS MORFOLOGICOS
## 5. Operaciones Morfológicas
- **Tipos**: Erosión (erode), dilatación (dilate), apertura (opening), cierre (closing), eliminación de agujeros (fill_holes), eliminación de objetos de diversos tamaños (fill)
- **Propósito**: Estas herramientas modifican la estructura de los objetos en una imagen binaria. La erosión reduce el tamaño de los objetos, mientras que la dilatación los expande. La apertura y el cierre son combinaciones de estas operaciones que se utilizan para eliminar ruido o rellenar huecos.
- **Importancia**: Las operaciones morfológicas son esenciales para limpiar imágenes binarias, separar objetos conectados y perfeccionar los límites de los objetos.

## 6. Métodos de transformación 
- **Tipos**: Rotar, redimensionar, autodetección de la tarjeta de color, fusión de imágenes, recortar, selección de región de inerés.
- **Propósito**: Permiten adecurar la imagen en el flujo de trabajo

## 7. Ecualización de Histograma
- **Tipos**: hist_equalization
- **Propósito**: Esta técnica ajusta el contraste de una imagen redistribuyendo las intensidades de los píxeles, de modo que se puedan mejorar las áreas de bajo contraste.
- **Importancia**: Ampliamente utilizada para mejorar el contraste global de una imagen, especialmente cuando el primer plano y el fondo son ambos brillantes o ambos oscuros, haciendo que las características sean más distinguibles.

## 8. Herramientas de Segmentación
- **Tipos**: Algoritmo de Watershed (watershed_segmentation), crecimiento de regiones, agrupamiento (por ejemplo, K-means (learn.train_kmeans), naive bayes (naive_bayes) )
- **Propósito**: La segmentación divide una imagen en regiones u objetos de interés, ayudando a aislar componentes para su análisis individual.
- **Importancia**: Fundamental para identificar y analizar objetos o regiones distintas, particularmente en imágenes médicas (por ejemplo, detección de tumores) o en la clasificación de objetos basada en imágenes.

## 9. Herramientas de Segmentación de estructuras vegetales
- **Tipos**: Funciones morfológicas: Enconrar ramas, encontrar puntas, encontrar ángulos de inserción de hojas,  
- **Propósito**: Describir la arquitectura de una planta
- **Importancia**: En fitomejoramiento y en fisiología vegetal.

## 9. Filtros de Deconvolución
- **Propósito**: Estos filtros se utilizan para revertir los efectos de desenfoque que ocurren durante la captura de imágenes debido a imperfecciones de la lente o movimiento.
- **Importancia**: La deconvolución es particularmente importante en la microscopía o en la imagen astronómica, donde los detalles finos son necesarios y las imágenes a menudo sufren varios tipos de desenfoque.

## 10. Filtros de Reducción de Ruido
- **Tipos**: Filtro de medios no locales, filtro bilateral, filtro de Wiener
- **Propósito**: Estos filtros están diseñados para eliminar diferentes tipos de ruido, como el ruido gaussiano, de sal y pimienta, o el ruido speckle, sin afectar demasiado la señal subyacente.
- **Importancia**: La reducción de ruido es esencial para mejorar la calidad de la imagen y garantizar que el ruido no interfiera con la detección y el análisis de características.

## 11. Transformaciones de Espacios de Color
- **Tipos**: De RGB a escala de grises, de RGB a HSV, espacio de color Lab
- **Propósito**: Transformar imágenes en diferentes espacios de color puede simplificar tareas como la detección de objetos o la segmentación basada en la información de color.
- **Importancia**: Ciertas tareas de análisis de imágenes, como la detección de piel o la identificación de objetos específicos, se vuelven más fáciles en un espacio de color diferente al estándar RGB.

## Importancia de Estas Herramientas en el Análisis de Imágenes
- **Mejora de la Calidad de la Imagen**: Los filtros y herramientas pueden mejorar la claridad visual de las imágenes, haciendo que las características clave sean más distinguibles.
- **Preprocesamiento para Aprendizaje Automático**: Muchas tareas de análisis de imágenes implican pasos de preprocesamiento, como suavizado, reducción de ruido y mejora del contraste, antes de introducir los datos en modelos de aprendizaje automático.
- **Detección de Objetos y Extracción de Características**: La detección de bordes, la umbralización y la segmentación permiten la identificación de objetos específicos y características, cruciales para tareas como los diagnósticos médicos o el control de calidad automatizado.
- **Reducción de Ruido**: Es esencial para extraer información significativa de datos con ruido, lo que es común en campos como la teledetección o la microscopía.
- **Automatización**: Los filtros y herramientas permiten flujos de trabajo

### APLICAR UMBRALIZACIÓN A LA IMAGEN EN ESCALA DE GRISES PARA BINARIZARLA

La umbralización es una técnica de procesamiento de imágenes que se utiliza para separar elementos específicos de una imagen, 
generalmente en blanco y negro. Consiste en convertir una imagen en escala de grises a una imagen binaria (blanco y negro) mediante la aplicación de un valor umbral (threshold) determinado.

#### ¿Cómo funciona?
- Seleccionar un umbral: Se elige un valor de umbral, normalmente entre 0 y 255, en función de la escala de grises de la imagen.
- Comparación de píxeles: Cada píxel de la imagen se compara con este valor de umbral:
    - Si el valor del píxel es mayor que el umbral, se convierte en blanco (1).
    - Si es menor o igual al umbral, se convierte en negro (0).

#### Aplicaciones comunes
- Segmentación de objetos: Para identificar y separar objetos en imágenes.
- Análisis de imágenes científicas: En microscopía y otras disciplinas, se utiliza para resaltar características específicas en las imágenes.
- Reconocimiento de patrones: Es un paso inicial en muchos algoritmos de reconocimiento de caracteres o detección de bordes.

La umbralización puede ser sencilla (con un solo umbral) o más compleja, utilizando técnicas adaptativas o basadas en histograma para elegir automáticamente el mejor valor de umbral según las características de la imagen.


```python
# Entradas:
#   gray_img - Imagen en escala de grises
#   threshold - Valor de umbral (0-255)
#   object_type - "light" o "dark" (por defecto: "light"). 
#                 Si el objeto es más claro que el fondo, se aplica un umbral estándar.
#                 Si el objeto es más oscuro que el fondo, se aplica un umbral inverso.

## Buscar la función: threshold.binary y aplicarla
bin_image=pcv.threshold.binary(gray_img=a_channel, threshold=115, object_type="dark")
```

```python
# Aplicar el método de umbralización de Otsu para binarizar la imagen.
# Otsu es un método automático que determina el umbral óptimo para separar los objetos del fondo, minimizando la varianza intraclase.

# otsu_image: Imagen binarizada donde los píxeles se clasifican en dos clases: fondo y objeto.
# gray_img: Imagen en escala de grises en la que se aplicará el umbral.
# object_type: Define el tipo de objeto para la binarización. En este caso, "dark" significa que el objeto es oscuro y el fondo es claro.
otsu_image=pcv.threshold.otsu(gray_img=a_channel, object_type="dark")
```

```python
# Aplicar umbralización adaptativa usando el método Gaussiano.
# El método Gaussiano ajusta el umbral de manera local en función de la intensidad de los píxeles cercanos.
# Es útil cuando hay variaciones en la iluminación a lo largo de la imagen.

# bin_gauss1: Imagen binarizada donde se aplica el umbral adaptativo basado en el método Gaussiano.
# gray_img: Imagen en escala de grises sobre la que se aplica el umbral.
# ksize: Tamaño del kernel que se utilizará en el filtro Gaussiano para calcular el umbral.
# offset: Valor que se agrega al resultado del filtro Gaussiano para ajustar el umbral.
# object_type: Define el tipo de objeto a identificar. "dark" indica que los objetos oscuros se deben separar del fondo claro.

bin_gauss1 = pcv.threshold.gaussian(gray_img=a_channel,
                                    ksize=1000,
                                    offset=2,
                                    object_type='dark')
```

```python
# Visualiza todos los colores para decidir cómo establecer el umbral.
# Se genera un gráfico de dispersión de los píxeles de la imagen para ayudar a identificar cómo se distribuyen los colores
# en los dos canales seleccionados (en este caso, 'a' y 'b' del espacio de color Lab) y cómo podrían usarse para segmentar.

# color_scatter: Gráfico de dispersión que visualiza la relación entre los dos canales 'a' y 'b'.
# paths_to_imgs: Lista de rutas a las imágenes que se van a analizar. En este caso, la imagen de la papa se encuentra en "./potato.jpg".
# x_channel: Canal que se usará para el eje X del gráfico (en este caso, el canal 'a' del espacio de color Lab).
# y_channel: Canal que se usará para el eje Y del gráfico (en este caso, el canal 'b' del espacio de color Lab).
    
color_scatter = pcv.visualize.pixel_scatter_plot(paths_to_imgs = ["./potato.jpg"],
                                                 x_channel = "a",
                                                 y_channel ="b" )
```

```python
# Aplicar umbralización en dos canales del espacio de color para segmentar la imagen.
# Esta función permite establecer un rango de valores de los canales 'a' y 'b' para la segmentación.

# thresh1: Imagen binarizada donde se establece un umbral en función de los canales 'a' y 'b' del espacio de color Lab.
# rgb_img: Imagen en formato RGB sobre la que se va a aplicar el umbral.
# x_channel: Canal 'a' del espacio de color Lab, que representa la información de color entre verde y rojo.
# y_channel: Canal 'b' del espacio de color Lab, que representa la información de color entre azul y amarillo.
# points: Coordenadas de dos puntos que definen el rango de valores de los canales para el umbral.
# above=True: Los píxeles que están por encima de los valores especificados en 'points' se consideran parte del objeto a segmentar.

# thresh1 = pcv.threshold.dual_channels(rgb_img = img, x_channel = "a", y_channel = "b", points = [(100,140),(130,180)], above=True)
thresh1 = pcv.threshold.dual_channels(rgb_img = img, 
                                      x_channel = "a", 
                                      y_channel = "b", 
                                      points = [(120,130),(130,180)], 
                                      above=True)
```

```python
# Rellenar los objetos pequeños en una imagen binarizada.
# Esta función se utiliza para eliminar áreas pequeñas (ruido) y rellenar los huecos en las regiones de interés.

# a_fill_image1: Imagen binarizada después de rellenar los objetos pequeños y eliminar los huecos dentro de las regiones de interés.
# bin_img: Imagen binarizada sobre la que se va a realizar el relleno.
# size: Tamaño mínimo de los objetos que se desean mantener. Los objetos más pequeños que este valor se eliminarán.

a_fill_image1 = pcv.fill(bin_img=thresh1, 
                         size=100)
```

```python
# Rellenar los huecos en los objetos binarizados.
# Esta función se utiliza para completar cualquier hueco o espacio vacío dentro de los objetos de la imagen binarizada.

# a_fill_image2: Imagen binarizada después de rellenar los huecos dentro de los objetos, eliminando vacíos internos.
# a_fill_image1: Imagen binarizada procesada previamente en la que se rellenaron los objetos pequeños.
# Esta función se asegura de que las regiones de interés no tengan agujeros o espacios vacíos internos.

a_fill_image2 = pcv.fill_holes(a_fill_image1)
```

```python
# Filtrar la máscara binaria para conservar solo las regiones que se superponen con una región de interés (ROI).
# Esta función permite aplicar una máscara binaria a una región específica de la imagen, manteniendo solo las áreas de la máscara que se encuentran dentro del ROI.

# kept_mask: Máscara binaria resultante, donde se conservan solo las áreas que se superponen parcialmente con la región de interés (ROI).
# mask: Máscara binaria sobre la que se aplicará el filtro. En este caso, es la imagen procesada "a_fill_image1".
# roi: La región de interés (ROI) que se definió previamente. En este caso, es un rectángulo que se creó con las coordenadas (x, y, h, w).
# roi_type='partial': Tipo de superposición. 'partial' significa que se conservarán solo las áreas de la máscara que se superponen parcialmente con el ROI.

kept_mask  = pcv.roi.filter(mask=a_fill_image1, roi=rect_roi, roi_type='partial')
```

```python
# Aplicar un filtro de mediana para suavizar la máscara binaria.
# El filtro de mediana es útil para reducir el ruido en las imágenes binarizadas sin borrar detalles importantes.

# median_img: Imagen binarizada suavizada después de aplicar el filtro de mediana.
# gray_img: La imagen binarizada sobre la que se aplicará el filtro. En este caso, es la máscara obtenida en el paso anterior "kept_mask".
# ksize: Tamaño del kernel que se utilizará en el filtro. El valor 13 representa un kernel de 13x13 píxeles. 
# El tamaño del kernel debe seleccionarse con cuidado para equilibrar la eliminación de ruido y la preservación de detalles.

median_img=pcv.median_blur(gray_img=kept_mask, ksize=13)

# Nota: El tamaño del kernel (ksize) debe elegirse cuidadosamente. Un kernel demasiado grande
# puede suavizar demasiado la imagen y perder detalles importantes, mientras que un kernel
# demasiado pequeño puede no eliminar suficiente ruido.
```

### APLICAR LA MÁSCARA A LA IMAGEN ORIGINAL

Aplicar una máscara a una imagen RGB

- plantcv.apply_mask(img, mask, mask_color)mage

Parametros:
- img: La imagen original (RGB o en escala de grises).
- mask: Imagen binaria donde las áreas en blanco (255) se conservarán y las negras (0) serán ocultadas.
- mask_color: 'white' o 'black', indica con qué color se rellenarán las áreas enmascaradas.

Contexto:
- Aplicar una máscara binaria sobre una imagen en escala de grises o RGB. 
- Útil para separar las plantas y los materiales de fondo. materials.

```python
# Aplicar la máscara binaria sobre la imagen original para segmentar las áreas de interés.
# Esta función toma la máscara procesada (en este caso, suavizada con el filtro de mediana) y la aplica sobre la imagen original.

# image_res: Imagen resultante después de aplicar la máscara. Solo se conservan las áreas de la imagen donde la máscara es blanca (representando la región de interés).
# img: Imagen original sobre la que se aplicará la máscara. En este caso, es la imagen completa que contiene la planta o el objeto de interés.
# mask: Máscara binaria que se aplicará. En este caso, es la imagen binarizada y suavizada con el filtro de mediana "median_img".
# mask_color='white': El color de la máscara. En este caso, se asume que las regiones de interés en la máscara son blancas y el fondo es negro.

image_res = pcv.apply_mask(img=img, mask=median_img, mask_color='white')
```

```python
# Aplicar una erosión morfológica a la imagen binaria para eliminar ruido y reducir regiones pequeñas.
# La erosión ayuda a eliminar píxeles blancos aislados y hace que los objetos se encojan.

# erode_img: Imagen resultante después de la erosión.
# gray_img: Imagen de entrada (debe ser binaria o en escala de grises).
# ksize: Tamaño del kernel (estructura) utilizado para la erosión. Un valor mayor erosiona más.
# i: Número de iteraciones de la erosión. Más iteraciones aplican el proceso varias veces.

erode_img = pcv.erode(gray_img=median_img, 
                      ksize=13, 
                      i=1)
```

```python
# Aplicar una dilatación morfológica a la imagen para expandir las regiones blancas.
# La dilatación es útil para rellenar huecos pequeños y restaurar partes de los objetos después de la erosión.

# dilate_img: Imagen resultante después de la dilatación.
# gray_img: Imagen de entrada (generalmente una imagen binaria o en escala de grises).
# ksize: Tamaño del kernel utilizado para la dilatación. Un valor mayor expande más las regiones blancas.
# i: Número de iteraciones de la dilatación. Más iteraciones aplican el proceso varias veces.

dilate_img = pcv.dilate(gray_img=erode_img, ksize=13, i=1)
```

```python
# Aplicar una máscara binaria sobre la imagen original para ocultar áreas no deseadas.
# Esto permite conservar solo las regiones de interés después del procesamiento morfológico.

# image_res: Imagen resultante después de aplicar la máscara.
# img: Imagen original (puede ser RGB o en escala de grises).
# mask: Máscara binaria donde los píxeles blancos (255) se conservarán y los negros (0) serán ocultados.
# mask_color: Define el color de fondo después de aplicar la máscara ('white' para fondo blanco, 'black' para fondo negro).

image_res=pcv.apply_mask(img=img, 
                         mask=dilate_img,
                         mask_color='white')
```

```python
# Guardar la imagen procesada en formato JPG.
# cv2.imwrite guarda la imagen en el archivo especificado.
# 'image_res.jpg': Nombre del archivo de salida en formato JPG.
# image_res: Imagen procesada que se va a guardar.
cv2.imwrite('image_res.jpg', image_res)

# Guardar la misma imagen en formato PNG.
# PNG es útil si se necesita una imagen sin pérdida de calidad o con transparencia.
cv2.imwrite('image_res.png', image_res)
```
